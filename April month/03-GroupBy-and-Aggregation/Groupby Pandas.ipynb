```python
import pandas as pd

data = {
    "Name": ["A","B","C","D","E","F","G","H","I","J"],
    "Sex": ["F","M","F","F","M","M","F","M","F","M"],
    "Pclass": [1,3,3,1,3,2,2,3,1,2],
    "Age": [38,22,26,35,35,28,19,40,30,25],
    "Fare": [71,7,8,53,8,13,26,7,80,15],
    "Survived": [1,0,1,1,0,0,1,0,1,0],
    "Embarked": ["C","S","S","S","S","Q","C","S","C","Q"],
    "FamilySize": [2,2,1,2,1,3,2,1,3,1]
}

df = pd.DataFrame(data)
df
```

```python
'''LEVEL 7 (REAL DATA SCIENCE THINKING)
Q16. Create:

👉 "Group_survival"

Where:

group = (Sex + Pclass)'''

df['Group_survival'] = df.groupby(['Pclass', 'Sex'])['Survived'].transform('mean')
df.head()
```

```python
df.groupby(['Sex', 'Pclass'])['Survived'].mean().sort_values(ascending=False)
```

```python
df.groupby(['Pclass'])['Fare'].std().sort_values(ascending = False)
```

```python
df['Family_survival_rate'] = df.groupby(['Sex', 'FamilySize'])['Survived'].transform('mean')
df.sample(4, replace = True)
```

```python
'''Q20. Create:

👉 IsAlone (0/1) → compare survival rate'''
df['isAlone'] = (df['FamilySize'] == 1).astype(int)
df.groupby('isAlone')['Survived'].mean()
```
